```python
import os
import requests
from edgar import *
import pandas as pd
from tqdm import tqdm

set_identity("ansa ansa1019@gmail.com")
headers = {
    "User-Agent": "NPUST@wp.npust.edu.tw",
    "Accept-Encoding": "gzip",
    "Host": "www.sec.gov",
}
df = pd.read_csv("sp500_companies.csv")
companies = df["Symbol"].values
```

```python
res = {}
lis = []
log = ""

root = "htm/"
if not os.path.isdir(root):
    os.makedirs(root)

for y in range(2014, 2025):
    folder = root + str(y) + "/"
    if not os.path.isdir(folder):
        os.makedirs(folder)

for y in range(2014, 2025):
    folder = "mda/" + str(y) + "/"
    if not os.path.isdir(folder):
        os.makedirs(folder)

for y in range(2014, 2025):
    folder = "pa/" + str(y) + "/"
    if not os.path.isdir(folder):
        os.makedirs(folder)

loop = tqdm(range(len(companies)))
for i in loop:
    error = ""
    try:
        company = Company(companies[i])

        filings = company.get_filings(form="10-K")
        if filings.empty:
            loop.set_description(
                f"{companies[i]}"
            )
            log += f"❌ {companies[i]} 無10-K可供下載\n"
            continue
        else:
            for filing in filings:
                if int(filing.report_date[:4]) >= 2014:
                    try:
                        path = root + filing.report_date[:4] + "/" + companies[i] + ".htm"
                        loop.set_description(
                            f"{companies[i]}_" +
                            filing.report_date[:4] + ".htm"
                        )
                        with open(path, "w") as f:
                            res = requests.get(
                                filing.primary_documents[0].url, headers=headers
                            )
                            f.write(res.text)
                    except Exception as e:
                        error += f"❌ {path} 下載失敗\n{e}"
                        continue
    except Exception as e:
        log += f"❌ {path} 下載失敗\n{e}"
        continue
    if error != "":
        log += error
    else:
        log += f"✅ {companies[i]} 下載完成\n"
print(log)
```

```python
res = {}
lis = []
log = ""

root = "mda/"
if not os.path.isdir(root):
    os.makedirs(root)

for y in range(2014, 2025):
    folder = root + str(y) + "/"
    if not os.path.isdir(folder):
        os.makedirs(folder)

loop = tqdm(range(len(companies)))
for i in loop:
    error = ""
    try:
        company = Company(companies[i])

        filings = company.get_filings(form="10-K").filter(date="2015-01-01:2024-12-31")
        if filings.empty:
            loop.set_description(
                f"{companies[i]}"
            )
            log += f"❌ {companies[i]} 無10-K可供下載\n"
            continue
        else:
            for filing in filings:
                try:
                    path = root + filing.report_date[:4] + "/" + companies[i] + ".htm"
                    loop.set_description(
                        f"{companies[i]}_" +
                        filing.report_date[:4] + ".htm"
                    )
                    with open(path, "w") as f:
                        tenk = filing.obj()
                        f.write(tenk["Item 7"])
                except Exception as e:
                    error += f"❌ {path} 下載失敗\n{e}"
                    continue
    except Exception as e:
        log += f"❌ {path} 下載失敗\n{e}"
        continue
    if error != "":
        log += error
    else:
        log += f"✅ {companies[i]} 下載完成\n"
print(log)
```
