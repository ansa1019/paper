```python
import os
from tqdm import tqdm
from sentence_transformers import util

# ‰∏ªÁ®ãÂ∫è
define_folder = "def/"
mda_folder = "mda/"

define_files = [f for f in os.listdir(define_folder) if f.endswith(".txt")]
mda_files = [f for f in os.listdir(mda_folder) if f.endswith(".txt")]

loop = tqdm(range(len(mda_files)))
log = ""
similarity = {}

# üîπ ‰øÆÊîπ `compare_texts()`ÔºåÁßªÈô§ÊªëÂãïË¶ñÁ™ó
def compare_texts(define_text, mda_text, model):
    """ Ë®àÁÆó define Ëàá mda ‰πãÈñìÁöÑÁõ∏‰ººÂ∫¶ """
    emb1 = model.encode(define_text)
    emb2 = model.encode(mda_text)
    return round(util.cos_sim(emb1, emb2).item(), 4)

for i in loop:
    mda_file = mda_files[i]
    sim = 0
    company_year = os.path.splitext(mda_file)[0]
    company, year = company_year.split("_")
    mda_file = f"{company_year}.txt"

    loop.set_description(mda_file)

    # üîπ ËÆÄÂèñÂÆåÊï¥ÁöÑ mda ÂÖßÂÆπ
    mda_path = os.path.join(mda_folder, mda_file)
    with open(mda_path, "r", encoding="utf-8") as file:
        mda_text = file.read()

    for define_file in define_files:
        define_path = os.path.join(define_folder, define_file)

        # üîπ ËÆÄÂèñÂÆåÊï¥ÁöÑ define ÂÖßÂÆπ
        with open(define_path, "r", encoding="utf-8") as file:
            define_text = file.read()

        # üîπ Áõ¥Êé•Ë®àÁÆó define Ëàá mda ÁöÑË™ûÊÑèÁõ∏‰ººÂ∫¶
        sim = compare_texts(define_text, mda_text, model)

    # üîπ Ë®òÈåÑ log
    log += f"{mda_file} Áõ∏‰ººÂ∫¶: {sim}\n"

    # üîπ Â∞áÁµêÊûúÂØ´ÂÖ• CSV
    if company not in similarity:
        similarity[company] = {}
    similarity[company][year] = sim

print(f"‚úÖ Áõ∏‰ººÂ∫¶Ë®àÁÆóÂÆåÊàê")
```
